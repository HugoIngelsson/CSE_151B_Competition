```python
import pandas as pd
import random
import json

# Load the Parquet file
df = pd.read_parquet('~/Downloads/train-00000-of-00005.parquet')

supplemental_training = []
for i in random.sample(list(range(0,len(df))), 4000):
    supplemental_training.append({'source': df['source'][i], 'problem': df['problem'][i], 
                'solution': df['solution'][i]})
        
print(len(supplemental_training))
with open('supplemental3.jsonl', 'w') as f:
    json.dump(supplemental_training, f)
```

```python
x, y = [], []
ravg, hist, hist_len = 0.0, [], 25
with open('lora_train_progress.txt', 'r') as f:
    for idx, line in enumerate(f):
        if idx % 4 == 0:
            x.append(int(line.strip()))

        elif idx % 4 == 1:
            hist.append(float(line.strip()))
            ravg += hist[-1]

            if len(hist) > hist_len:
                ravg -= hist[0]
                hist.pop(0)
            
            y.append(ravg / len(hist))
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
```
